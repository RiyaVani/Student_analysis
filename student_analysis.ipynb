```python
### VISUALIZATION
```

```python
## HISTOGARM OF GRADE
plt.hist(df['G3'],bins=9,color="pink")
plt.xlabel("Final grades (g3)")
plt.ylabel("Number of student")
plt.title("Histogram of final grades")
plt.show()

```

```python
## Scatter plot: Study Time vs grades

plt.scatter(df["studytime"],df['G3'],color= ['black', 'red'])
plt.xlabel("studytime")
plt.ylabel("Grades")
plt.title("studytime vs grades")
plt.show()
```

```python
### BAR CHART : MALE VS FEMALE AVEARGE SCORE

df.groupby("sex")["G3"].mean().plot.bar(color=["black", "pink"])
plt.xlabel("Gender")
plt.ylabel("Average final score")
plt.title("Avrage score ")
plt.show()
```
